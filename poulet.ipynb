```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
bilan = pd.read_csv("bilan_alimentaire.csv")
demog = pd.read_csv("demographie.csv")
macro = pd.read_csv("indicateur_macro.csv")
secu = pd.read_csv("securite_alimentaire.csv")
```

### bilan

```python
bilan_dum = pd.get_dummies(bilan.Élément).mul(bilan.Valeur, axis = "index")
bilan_dum = bilan_dum.join(bilan)
```

```python
bilan_dum.columns = bilan_dum.columns.str.replace(r"[()]", "")
bilan_dum.columns = bilan_dum.columns.str.replace(r"[/]", "")
bilan_dum.columns = bilan_dum.columns.str.replace(r"[-]", "")
bilan_dum.columns = bilan_dum.columns.str.replace(r"[,]", "")
bilan_dum.columns = bilan_dum.columns.str.replace(r"[&]", "")
```

```python
bilan_dum.set_index(["Zone",'Année'])
bilan_dum2 = bilan_dum.groupby(["Zone",'Année']).sum()
```

```python
bilan_dum2
```

### macro

```python
macro_dum = pd.get_dummies(macro.Élément).mul(macro.Valeur, axis = "index")
macro_dum = macro_dum.join(macro)
```

```python
macro_dum.columns = macro_dum.columns.str.replace(r"[()]", "")
macro_dum.columns = macro_dum.columns.str.replace(r"[/]", "")
macro_dum.columns = macro_dum.columns.str.replace(r"[-]", "")
macro_dum.columns = macro_dum.columns.str.replace(r"[,]", "")
macro_dum.columns = macro_dum.columns.str.replace(r"[&]", "")
```

```python
macro_dum.set_index(["Zone",'Année'])
macro_dum2 = macro_dum.groupby(["Zone",'Année']).sum()
```

### demog

```python
demog_dum = pd.get_dummies(demog.Élément).mul(demog.Valeur, axis = "index")
demog_dum = demog_dum.join(demog)
```

```python
demog_dum.columns = demog_dum.columns.str.replace(r"[()]", "")
demog_dum.columns = demog_dum.columns.str.replace(r"[/]", "")
demog_dum.columns = demog_dum.columns.str.replace(r"[-]", "")
demog_dum.columns = demog_dum.columns.str.replace(r"[,]", "")
demog_dum.columns = demog_dum.columns.str.replace(r"[&]", "")
```

```python
demog_dum2
```

### secu

```python
secu_dum = pd.get_dummies(secu.Produit,dtype = "int64").mul(secu.Valeur, axis = "index")

secu_dum = secu_dum.join(secu)
```

```python
secu_dum.columns = secu_dum.columns.str.replace(r"[()]", "")
secu_dum.columns = secu_dum.columns.str.replace(r"[/]", "")
secu_dum.columns = secu_dum.columns.str.replace(r"[-]", "")
secu_dum.columns = secu_dum.columns.str.replace(r"[,]", "")
secu_dum.columns = secu_dum.columns.str.replace(r"[&]", "")
```

```python
secu_dum = secu_dum.drop(["Domaine","Élément","Produit","Description du Symbole","Note","Symbole","Valeur","Unité","Code année","Code Produit","Code Élément","Code Élément","Code Élément","Code zone M49","Code Domaine"], axis = 1)
```

```python
secu_dum =  secu_dum.set_index(['Zone','Année'])
```

```python
secu_dum = secu_dum.replace("",np.nan)
secu_dum = secu_dum.replace("<0.1",0.0)
secu_dum = secu_dum.replace("<2.5", 2.499)
secu_dum = secu_dum.replace("<0.5", 0.499)
secu_dum = secu_dum.replace("FS",np.nan)
```

```python
secu_dum = secu_dum.astype("float")
```

```python
secu_dum2 = secu_dum.groupby(["Zone",'Année']).sum()
```

```python
secu_dum2
```

### final

```python
final = bilan_dum2.merge(macro_dum2, how="inner",on="Zone")
final = final.merge(demog_dum2, how= "inner", on="Zone")
final= final.merge(secu_dum2, how= "inner", on = "Zone")
```

```python
final = final.drop(["Valeur","Code année","Code Élément","Code Élément","Code Élément","Code zone M49","Code zone M49_x","Code Élément_x","Code année_x","Code zone M49_y","Code Élément_y","Code Produit_x","Code année_y","Note_x","Note_y","Code Produit_y","Valeur_x","Valeur_y"], axis = 1)
```
